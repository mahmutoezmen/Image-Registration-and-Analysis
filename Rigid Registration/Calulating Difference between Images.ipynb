```python
import numpy as np
from PIL import Image 
import nibabel as nib
%matplotlib inline 
from matplotlib import pyplot as plt
```

# For Affine Transformation

```python
fixed_image  = Image.open('fixed.png')
resulting_image = nib.load('Desktop/result.0.nii')
```

```python
fixed_image = np.asarray(fixed_image)
resulting_image = np.asarray(resulting_image.dataobj)
ima = np.abs(fixed_image.T-resulting_image)
print(np.sum(ima))
plt.imshow(ima.T, interpolation='nearest',cmap='gray')
plt.show()
```
